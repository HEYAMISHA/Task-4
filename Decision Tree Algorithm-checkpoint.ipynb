```python
# Import the required library.
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns
sns.set()
```

```python
# Now that we have sklearn imported in our notebook, we can begin working with the dataset for our machine learning model.
from sklearn.datasets import load_iris
iris = load_iris()
```

# Preparing the Data

```python
# We will now divide our data into attributes and labels and will then divide the resultant data into both training and test sets.
x = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2)
```

# Training and Making Predictions

```python
#  train the decision tree algorithm on this data and make predictions
traintree=DecisionTreeClassifier()
traintree.fit(x_train,y_train)
```

```python
# Make prediction
y_train
```

```python
y_hot = traintree.predict(x_train)#predictions are stored in y^(y_train should match y_hot)
y_hot
```

```python
# Comparing Actual vs Predicted
dt_performance = pd.DataFrame({'Actual_train': y_train, 'Predicted_train': y_hot})  
dt_performance
```

```python
y_pred = traintree.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Comparing Actual vs Predicted
dt_test= pd.DataFrame({'Actual_test': y_test, 'Predicted_test': y_pred})  
dt_test
```

# Evaluating the Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import plot_confusion_matrix, accuracy_score
from sklearn.svm import SVC
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# From the confusion matrix, you can see that out of 90 test instances, our algorithm misclassified only 6. This is 98.5 % accuracy. Not too bad!
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

```python
print('PREDICTION ACCURACY : \033[1;32m{}'.format(accuracy_score(y_test, y_pred)))
```
